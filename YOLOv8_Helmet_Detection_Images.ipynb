#**Loading Trained Model**

```python
path_project_folder = '/content/drive/MyDrive/Helmet Detection System'  ## change to the project folder path
model_path = os.path.join(path_project_folder,'runs/detect/train3/weights/best.pt')  ## provide the path to latest train folder
```

```python
trained_model = YOLO(model_path)
```

#**Input Test Image**
Load your test image below

```python
from google.colab import files
image_file = files.upload()
```

#**Processing Image & Predicting**




```python
image = cv2.imread(str(list(image_file.keys())[0]))

orig_height, orig_width , channels = image.shape

scaled_width = 640
scaled_height = 640
resized_image = cv2.resize(image,(scaled_width,scaled_height))
```

```python
results = trained_model.predict(resized_image,conf =0.5)
annotated_image = results[0].plot()
```

#**Output Image**

```python
reverted_ann_image = cv2.resize(annotated_image,(orig_width,orig_height))
cv2_imshow(reverted_ann_image)
```
